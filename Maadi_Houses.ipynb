```python
import requests
from bs4 import BeautifulSoup
import re 
import pandas as pd
```

```python
page_1=requests.get('https://housesolutionegypt.com/sale')
page_2=requests.get('https://housesolutionegypt.com/sale?page=2')
page_3=requests.get('https://housesolutionegypt.com/sale?page=3')
page_4=requests.get('https://housesolutionegypt.com/sale?page=4')
page_5=requests.get('https://housesolutionegypt.com/sale?page=5')
page_6=requests.get('https://housesolutionegypt.com/sale?page=6')
page_7=requests.get('https://housesolutionegypt.com/sale?page=7')
page_8=requests.get('https://housesolutionegypt.com/sale?page=8')
page_9=requests.get('https://housesolutionegypt.com/sale?page=9')

with open('page1.html',mode='wb') as page1:
    page1.write(page_1.content)
with open('page2.html',mode='wb') as page2:
    page2.write(page_2.content)
with open('page3.html',mode='wb') as page3:
    page3.write(page_3.content)
with open('page4.html',mode='wb') as page4:
    page4.write(page_4.content)
with open('page5.html',mode='wb') as page5:
    page5.write(page_5.content)
with open('page6.html',mode='wb') as page6:
    page6.write(page_6.content)
with open('page7.html',mode='wb') as page7:
    page7.write(page_7.content)
with open('page8.html',mode='wb') as page8:
    page8.write(page_8.content)
with open('page9.html',mode='wb') as page9:
    page9.write(page_9.content)
    
with open('page1.html',encoding='utf-8') as file:
    p1=BeautifulSoup(file)

with open('page2.html',encoding='utf-8') as file:
    p2=BeautifulSoup(file)
    
with open('page3.html',encoding='utf-8') as file:
    p3=BeautifulSoup(file)

with open('page4.html',encoding='utf-8') as file:
    p4=BeautifulSoup(file)
    
with open('page5.html',encoding='utf-8') as file:
    p5=BeautifulSoup(file)

with open('page6.html',encoding='utf-8') as file:
    p6=BeautifulSoup(file)
    
with open('page7.html',encoding='utf-8') as file:
    p7=BeautifulSoup(file)

with open('page8.html',encoding='utf-8') as file:
    p8=BeautifulSoup(file)
    
with open('page9.html',encoding='utf-8') as file:
    p9=BeautifulSoup(file)
```

```python
place=[]
properties=[]
area=[]
bathrooms=[]
finishing_level=[]
bedrooms=[]
price=[]
pages_list=[p1,p2,p3,p4,p5,p6,p7,p8,p9]
for _ in pages_list:
    
    for _1 in range(len(_.find_all('div','row p-0 text-white'))):
        
        place.append(str(_.find_all('div','col-xs-12 col-6 px-0')[_1].contents[1]).strip().replace('  ',''))
        
        properties.append(_.find_all('div','row font-12')[_1].contents[4].find('div','mb-1 text-center text-white rounded p-1 bg-main-color').contents[0])
        
        area.append(re.search(r'(\d+ Sqm)',str(_.find_all('div','row font-12')[_1].contents[0])).group())
        
        bathrooms.append(re.search(r'(\d+ Bathrooms)',str(_.find_all('div','row font-12')[_1].contents[2])).group())
        
        bedrooms.append(re.search(r'(\d+ Bedrooms)',str(_.find_all('div','row font-12')[_1].contents[1])).group())
        
        finishing_level.append(re.search(r'(\w+ (\w+)?)',_.find_all('div','row font-12')[_1].contents[3].find('div','mb-1 py-2 text-xs-center text-secondary').contents[1]).group())
        
        price.append(str(_.find_all('div','row p-0 text-white')[_1].find('div','col-xs-12 col-6 text-right px-0').contents[0]).strip())
```

```python
Maadi_houses=pd.DataFrame(
{
  'place':place,
  'properties':properties,
  'finishing_level':finishing_level,
  'area':area,
  'bathrooms':bathrooms,
  'bedrooms':bedrooms,
  'price':price,
}
)
```
